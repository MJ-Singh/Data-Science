```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
problem_data = pd.read_csv('problem_data.csv')
user_data = pd.read_csv('user_data.csv')
train_submissions = pd.read_csv('train_submissions.csv')

test_data = pd.read_csv('test_submissions_NeDLEvX.csv')
```

```python
train_problem = train_submissions.join(problem_data.set_index('problem_id'), on='problem_id')
test_data = test_data.join(problem_data.set_index('problem_id'), on='problem_id')
```

```python
train_data = train_problem.join(user_data.set_index('user_id'), on='user_id')
test_data = test_data.join(user_data.set_index('user_id'), on='user_id')
```

```python
tags_data = pd.DataFrame(train_data['tags'].str.split(',', 7, expand=True))
tags_data.columns = ['tag1','tag2','tag3','tag4','tag5','tag6','tag7']

tags_test = pd.DataFrame(test_data['tags'].str.split(',', 7, expand=True))
tags_test.columns = ['tag1','tag2','tag3','tag4','tag5','tag6','tag7']
```

```python
tags_data.replace([None], [0]) 
```

```python
tags_test.replace([None], [0]) 
```

```python
tag1 = pd.get_dummies(tags_data['tag1'])
tag2 = pd.get_dummies(tags_data['tag2'])
tag3 = pd.get_dummies(tags_data['tag3'])
tag4 = pd.get_dummies(tags_data['tag4'])
tag5 = pd.get_dummies(tags_data['tag5'])
tag6 = pd.get_dummies(tags_data['tag6'])
tag7 = pd.get_dummies(tags_data['tag7'])

testtag1 = pd.get_dummies(tags_test['tag1'])
testtag2 = pd.get_dummies(tags_test['tag2'])
testtag3 = pd.get_dummies(tags_test['tag3'])
testtag4 = pd.get_dummies(tags_test['tag4'])
testtag5 = pd.get_dummies(tags_test['tag5'])
testtag6 = pd.get_dummies(tags_test['tag6'])
testtag7 = pd.get_dummies(tags_test['tag7'])
```

```python
tag1['schedules'] = 0
testtag1['schedules'] = 0
```

```python
tag1[tag2.columns] = tag1[tag2.columns].add(tag2)
tag1[tag3.columns] = tag1[tag3.columns].add(tag3)
tag1[tag4.columns] = tag1[tag4.columns].add(tag4)
tag1[tag5.columns] = tag1[tag5.columns].add(tag5)
tag1[tag6.columns] = tag1[tag6.columns].add(tag6)
tag1[tag7.columns] = tag1[tag7.columns].add(tag7)

testtag1[testtag2.columns] = testtag1[testtag2.columns].add(testtag2)
testtag1[testtag3.columns] = testtag1[testtag3.columns].add(testtag3)
testtag1[testtag4.columns] = testtag1[testtag4.columns].add(testtag4)
testtag1[testtag5.columns] = testtag1[testtag5.columns].add(testtag5)
testtag1[testtag6.columns] = testtag1[testtag6.columns].add(testtag6)
testtag1[testtag7.columns] = testtag1[testtag7.columns].add(testtag7)
```

```python
data = pd.concat([train_data,tag1,pd.get_dummies(train_data['rank']),pd.get_dummies(train_data['level_type'])], axis=1)


last_data = pd.concat([test_data,testtag1,pd.get_dummies(test_data['rank']),pd.get_dummies(test_data['level_type'])], 
                      axis=1)
```

```python
data.drop(['tags','rank','level_type'],axis=1,inplace=True)
last_data.drop(['tags','rank','level_type'],axis=1,inplace=True)
```

```python
train_data.country = pd.Categorical(train_data.country)
data['country'] = train_data.country.cat.codes

test_data.country = pd.Categorical(test_data.country)
last_data['country'] = test_data.country.cat.codes

train_data.user_id = pd.Categorical(train_data.user_id)
data['user_id'] = train_data.user_id.cat.codes

test_data.user_id = pd.Categorical(test_data.user_id)
last_data['user_id'] = test_data.user_id.cat.codes

train_data.problem_id = pd.Categorical(train_data.problem_id)
data['problem_id'] = train_data.problem_id.cat.codes

test_data.problem_id = pd.Categorical(test_data.problem_id)
last_data['problem_id'] = test_data.problem_id.cat.codes
```

```python
data['country'][data['country'] == -1] = np.nan

last_data['country'][last_data['country'] == -1] = np.nan
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                                                 ,'follower_count', 'last_online_time_seconds', 'max_rating'
                                                 ,'rating', 'registration_time_seconds']])
data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                      ,'follower_count', 'last_online_time_seconds', 'max_rating'
                      ,'rating', 'registration_time_seconds']] = scaler.transform(data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                      ,'follower_count', 'last_online_time_seconds', 'max_rating'
                      ,'rating', 'registration_time_seconds']])

scaler2 = preprocessing.StandardScaler().fit(last_data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                                                 ,'follower_count', 'last_online_time_seconds', 'max_rating'
                                                 ,'rating', 'registration_time_seconds']])
last_data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                      ,'follower_count', 'last_online_time_seconds', 'max_rating'
                      ,'rating', 'registration_time_seconds']] = scaler2.transform(last_data[['user_id','problem_id','submission_count','problem_solved', 'contribution'
                      ,'follower_count', 'last_online_time_seconds', 'max_rating'
                      ,'rating', 'registration_time_seconds']])
```

```python
from sklearn import preprocessing
country = pd.DataFrame(data['country'][data['country'].notnull()])
scaler = preprocessing.StandardScaler().fit(country)
country = scaler.transform(country)

from sklearn import preprocessing
points = pd.DataFrame(data['points'][data['points'].notnull()])
scaler = preprocessing.StandardScaler().fit(points)
points = scaler.transform(points)


data['country'][data['country'].notnull()] = country

data['points'][data['points'].notnull()] = points


from sklearn import preprocessing
country = pd.DataFrame(last_data['country'][last_data['country'].notnull()])
scaler = preprocessing.StandardScaler().fit(country)
country = scaler.transform(country)

from sklearn import preprocessing
points = pd.DataFrame(last_data['points'][last_data['points'].notnull()])
scaler = preprocessing.StandardScaler().fit(points)
points = scaler.transform(points)


last_data['country'][last_data['country'].notnull()] = country

last_data['points'][last_data['points'].notnull()] = points
```

```python
newdata = data[data['attempts_range'] != 6]
newdata = newdata[newdata['attempts_range'] != 5]
newdata = newdata[newdata['attempts_range'] != 4]
newdata = newdata[newdata['attempts_range'] != 3]
```

```python
y = newdata['attempts_range']
newdata.drop(['attempts_range'], axis=1, inplace=True)

ID = last_data['ID']
last_data.drop(['ID'], axis=1, inplace=True)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(newdata, y, test_size=0.20)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
from xgboost import XGBClassifier

xgbt = XGBClassifier(learning_rate =0.25,
                     n_estimators=250,
                     max_depth=11,
                     min_child_weight=6,
                     gamma=0.1,
                     reg_alpha=0.01,
                     subsample=0.8,
                     colsample_bytree=0.8,
                     objective= 'binary:logistic',
                     scale_pos_weight=1,
                     seed=27)
xgbt.fit(X_train, y_train)
pred = xgbt.predict(X_test)
print(xgbt.score(X_train, y_train))
print(xgbt.score(X_test, y_test))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, pred, average='weighted') 
```
