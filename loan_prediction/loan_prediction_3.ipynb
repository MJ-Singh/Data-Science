```python
#Data analysis
import numpy as np
import pandas as pd
from pandas import DataFrame,Series

#Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cross_validation import train_test_split
```

```python
#load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head(10)
```

```python
def num_missing(x):
    return sum(x.isnull())
train.apply(num_missing, axis=0)
```

```python
print(train.info())
print('-'*40)
print(test.info())
```

```python
train = train.drop("Loan_ID",axis=1)
loan_id = test['Loan_ID']
test = test.drop("Loan_ID",axis=1)
train.describe(include='all')
```

```python
from scipy.stats import mode

train['Gender'].fillna(train['Gender'].mode()[0], inplace=True)
train['Married'].fillna(train['Married'].mode()[0], inplace=True)
train['Dependents'].fillna(train['Dependents'].mode()[0], inplace=True)
train['Self_Employed'].fillna(train['Self_Employed'].mode()[0], inplace=True)
#train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mode()[0], inplace=True)

test['Gender'].fillna(test['Gender'].mode()[0], inplace=True)
test['Married'].fillna(test['Married'].mode()[0], inplace=True)
test['Dependents'].fillna(test['Dependents'].mode()[0], inplace=True)
test['Self_Employed'].fillna(test['Self_Employed'].mode()[0], inplace=True)
#test['Loan_Amount_Term'].fillna(test['Loan_Amount_Term'].mode()[0], inplace=True)
train.apply(num_missing, axis=0)
```

```python
#Determine pivot table
loan_amount = train.pivot_table(values=["LoanAmount"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(loan_amount)

credit_history = train.pivot_table(values=["Credit_History"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(credit_history)

loan_amount_term = train.pivot_table(values=["Loan_Amount_Term"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(loan_amount_term)

loan_amount_test = test.pivot_table(values=["LoanAmount"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(loan_amount_test)

credit_history_test = test.pivot_table(values=["Credit_History"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(credit_history_test)

loan_amount_term_test = test.pivot_table(values=["Loan_Amount_Term"], index=["Gender","Married","Self_Employed"], aggfunc=np.mean)
print(loan_amount_term_test)
```

```python
#iterate only through rows with missing column
for i,row in train.loc[train['LoanAmount'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    train.loc[i,'LoanAmount'] = loan_amount.loc[ind].values[0]

for i,row in train.loc[train['Credit_History'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    train.loc[i,'Credit_History'] = credit_history.loc[ind].values[0]

for i,row in train.loc[train['Loan_Amount_Term'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    train.loc[i,'Loan_Amount_Term'] = loan_amount_term.loc[ind].values[0]
    
for i,row in test.loc[test['LoanAmount'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    test.loc[i,'LoanAmount'] = loan_amount_test.loc[ind].values[0]

for i,row in test.loc[test['Credit_History'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    test.loc[i,'Credit_History'] = credit_history_test.loc[ind].values[0]

for i,row in test.loc[test['Loan_Amount_Term'].isnull(),:].iterrows():
    ind = tuple([row['Gender'],row['Married'],row['Self_Employed']])
    test.loc[i,'Loan_Amount_Term'] = loan_amount_term_test.loc[ind].values[0]

#Now check the #missing values again to confirm:
print(train.apply(num_missing, axis=0))
print(test.apply(num_missing, axis=0))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train['Loan_Status'] = encoder.fit_transform(train['Loan_Status'].astype('str'))
train.describe(include='all')
```

```python
test.describe(include='all')
```

```python
train.describe(include='all')
train = train = pd.get_dummies(train)
test = test = pd.get_dummies(test)
```

```python
train.head()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']] = scaler.fit_transform(train[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']])
```

```python
test.head()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
test[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']] = scaler.fit_transform(test[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']])
```

```python
y = train['Loan_Status']
X = train.drop(['Loan_Status'],axis=1)
X.head()
```

```python
test.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
X_train.head()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm = GradientBoostingClassifier(n_estimators=20, learning_rate=.07,
                                max_depth=2, random_state=0, subsample=0.8, max_features=0.8)
gbm.fit(X_train, y_train)
pred = gbm.predict(X_test)
print(gbm.score(X_train, y_train))
print(gbm.score(X_test, y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2), n_estimators=20, 
                         learning_rate=.06, algorithm="SAMME", random_state=0)
ada.fit(X_train, y_train)
pred = ada.predict(X_test)
print(ada.score(X_train, y_train))
print(ada.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators= 20, oob_score = True, n_jobs = -1,random_state =17, max_features = 0.9)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
print(rf.score(X_train, y_train))
print(rf.score(X_test, y_test))
```
